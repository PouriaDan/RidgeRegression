```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from ridge_regression import RidgeRegression
```

```python
%matplotlib inline
X = np.linspace(0,1,250).reshape(-1,1)
noise = (np.random.rand(250,)*0.5).reshape(-1,1)
Y = 3*X-0.5+noise

rr = RidgeRegression()
rr.fit(X,Y)
preds = rr.predict(X)

fig, ax = plt.subplots()
fig.set_size_inches(6,4)
ax.scatter(X,Y, s=3, label='Ground Truth')
ax.scatter(X, preds, s=3, label='Predictions')
ax.legend()
plt.show()
```

```python
%matplotlib inline
X = np.linspace(0,1,250).reshape(-1,1)
noise = (np.random.rand(250,)*0.5).reshape(-1,1)
Y = 10*X**2-7*X+0.5+noise

rr = RidgeRegression(normalize=True)
features = np.concatenate([X**i for i in range(1,2+1)], axis=1)
rr.fit(features,Y)
preds = rr.predict(features)

fig, ax = plt.subplots()
ax.scatter(X,Y, s=3, label='Ground Truth')
ax.scatter(X, preds, s=3, label='Predictions')
ax.legend()
plt.show()
```
